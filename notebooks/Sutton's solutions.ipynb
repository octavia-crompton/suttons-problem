### Computing $E_a$

Equation (7.45) expresses a *constant* vertical turbulent moisture flux:
$$
E_a \;=\; -\,\rho\,K_v(z)\,\frac{\partial  q(z)}{\partial z}
\;=\;\text{const.}
\tag{7.45}
$$

Assume a power-law eddy diffusivity,
$$
K_v(z)=b\,z^n,\qquad n=1-m
$$

Substituting into (7.45) and integrating from $z_s$ to $z_a$ gives
$$
 q(z_a)- q(z_s)
= -\frac{E_a}{\rho}\int_{z_s}^{z_a}\frac{dz}{b z^n}
= -\frac{E_a}{\rho b}\,\frac{z_a^{1-n}-z_s^{1-n}}{1-n},
$$

so
$$
E_a
=\rho\,b\,(1-n)\,\frac{q_s-q_a}{z_a^{\,1-n}-z_s^{\,1-n}},
$$

with $q_s=0.622\,e_s/p$ and $q_a=0.622\,e_a/p$.


With $n=1-m$ (so $1-n=m$) and
$$
b=\frac{u_*\,z_0^m}{C_p\,m},
$$
the flux can also be written as
$$
E_a
=\rho\,\frac{u_*\,z_0^m}{C_p}\,\frac{q_s-q_a}{z_a^m-z_s^m}
=\rho\,\frac{u_*\,z_0^m}{C_p}\,\frac{0.622}{p}\,\frac{e_s-e_a}{z_a^m-z_s^m}.
$$


#### More details

Specific humidity at the two heights:
$$
q_s = 0.622\,\frac{e_s}{p},
\qquad
q_a = 0.622\,\frac{e_a}{p}.
$$

Power-law wind profile used to infer friction velocity:
$$
U(z) = C_p\,u_* \left(\frac{z}{z_0}\right)^m,
\qquad
u_* = {U(z_{\mathrm{ref}})}C_p^{-1} \left(\dfrac{z_{\mathrm{ref}}}{z_0}\right)^{-m},
\qquad
z_{\mathrm{ref}} = 200~\mathrm{cm}.
$$

Eddy diffusivity form:
$$
K_v(z) = b\,z^n, \qquad n = 1-m,
\qquad
b = \frac{u_*\,z_0^{m}}{C_p\,m}.
$$

Starting from Eq. (7.54),
$$
{E}
 = E_a + \rho b\left(\frac{a}{b x_f}\right)^{\nu}
   \frac{(1-\nu)^{2\nu-2}(m+1)^{1-2\nu}}{\Gamma(\nu)}
   \left({q}_s-{q}_{as}\right),
\tag{7.54}
$$
divide through by $E_a$ to obtain
$$
\frac{{E}-E_a}{E_a}
=
\frac{\rho b}{E_a}\left(\frac{a}{b x_f}\right)^{\nu}
\frac{(1-\nu)^{2\nu-2}(m+1)^{1-2\nu}}{\Gamma(\nu)}
\left({q}_s-{q}_{as}\right).
\tag{1}
$$

$E_a$ computation from Eq. (7.45) with $n=1-m$:

$$
E_a=\rho\,b\,m\,\frac{q_s-q_a}{z_a^{m}-z_s^{m}}
\quad\Rightarrow\quad
\frac{\rho b}{E_a}=\frac{z_a^{m}-z_s^{m}}{m\,(q_s-q_a)}.
\tag{2}
$$

Substitute (2) into (1):
$$
\frac{{E}-E_a}{E_a}
=
\frac{z_a^{m}-z_s^{m}}{m}
\left(\frac{a}{b x_f}\right)^{\nu}
\frac{(1-\nu)^{2\nu-2}(m+1)^{1-2\nu}}{\Gamma(\nu)}
\tag{3}
$$

Using Eq. (7.55) (with $C_p=5.5$):
$$
a=\frac{C_p u_*}{z_s^m},\qquad b=\frac{u_* z_s^m}{C_p m}
\;\Rightarrow\;
\left(\frac{a}{b x_f}\right)^\nu
=
\left(\frac{C_p^2 m}{z_s^{2m}x_f}\right)^\nu.
\tag{4}
$$



```python
import numpy as np
import matplotlib.pyplot as plt
from math import gamma

# ---  defaults (Brutsaert Ch. 7 example conditions) ---
p_mb  = 1013.2          # pressure [mb]
rho   = 1.2e-3          # air density [g/cc] (cubic centimeter) 
z2_cm = 200.0           # 2 m = 200 cm

def q_from_e(e_mb, p_mb=p_mb):
    """
    Specific humidity (kg/kg, dimensionless) from vapor pressure (mb).
    """
    return 0.622 * e_mb / p_mb

def ustar_from_u2(u2_mps, z0_cm, m, Cp=5.5):
    """
    Invert U(z) = Cp*u_*(z/z0)^m at z=2 m to get u_*.
    Cp is a parameter (equation 4.19)
    Returns u_* in cm/s 
    (consistent with z0 in cm, z in cm).
     """
    
    u2_cms = 100.0 * u2_mps
    
    return u2_cms / (Cp * (z2_cm / z0_cm) ** m)

def ab_from_ustar(ustar_cms, z0_cm, m, Cp=5.5):
    
    """Eq. (7.55): 
    a = Cp*u_*/z0^m and 
    b = u_* z0^m/(Cp*m)."""
    
    a = Cp * ustar_cms / (z0_cm ** m)
    b = ustar_cms * (z0_cm ** m) / (Cp * m)
    return a, b

def calculate_parameters(xf_m, u2_mps, de_mb,
                         m=1/7, 
                         z0_cm=0.03, 
                         Cp=5.5,
                         rho=rho, p_mb=p_mb):
    """
    Compute (Ebar - Ea) from Eq. (7.54) as a function of fetch xf_m [m],
    2-m wind u2_mps [m/s], 
    and vapor-pressure jump de_mb = (es - eas) [mb].

    Returns enhancement in g cm^-2 s^-1 
    (multiply by 864000 to get mm/day).
    """
    # exponents: n=1-m (Eq 7.19), v = m/(1+2m)
    n = 1 - m
    v = (1 - n)/(2 + m - n)   # == m/(1+2m)

    # convert vapor-pressure jump -> specific humidity jump
    dq = q_from_e(de_mb, p_mb)

    # u2 -> u* -> (a,b)
    ustar = ustar_from_u2(u2_mps, z0_cm, m, Cp = Cp)

    a, b = ab_from_ustar(ustar, z0_cm, m, Cp = Cp)

    xf_cm = 100.0 * xf_m


    pref = (1 - v)**(2*v - 2) * (m + 1)**(1 - 2*v)
    term = (a**v) * (b**(1 - v)) * (xf_cm**(-v)) * pref / gamma(v)

    return v, a, b



def evap_enhancement_754(xf_m, u2_mps, de_mb,
                         m=1/7, 
                         z0_cm=0.03, 
                         Cp=5.5,
                         rho=rho, p_mb=p_mb):
    """
    Compute (Ebar - Ea) from Eq. (7.54) as a function of fetch xf_m [m],
    2-m wind u2_mps [m/s], 
    and vapor-pressure jump de_mb = (es - eas) [mb].

    Returns enhancement in g cm^-2 s^-1 
    (so multiply by 864000 to get mm/day).
    """
    # exponents: n=1-m (Eq 7.19), v = m/(1+2m)
    n = 1 - m
    v = (1 - n)/(2 + m - n)   # == m/(1+2m)

    # convert vapor-pressure jump -> specific humidity jump
    dq = q_from_e(de_mb, p_mb)

    # u2 -> u* -> (a,b)
    ustar = ustar_from_u2(u2_mps, z0_cm, m, Cp = Cp)

    a, b = ab_from_ustar(ustar, z0_cm, m, Cp = Cp)

    xf_cm = 100.0 * xf_m


    pref = (1 - v)**(2*v - 2) * (m + 1)**(1 - 2*v)
    term = (a**v) * (b**(1 - v)) * (xf_cm**(-v)) * pref / gamma(v)

    return rho * dq * term

from math import log

def Ea_from_745_powerlaw(u2_mps, z0_cm, m,
                         e1_mb, e2_mb, z1_cm, z2_cm,
                         Cp=5.5, rho=1.2e-3, p_mb=1013.2):
    """
    Compute E_a from Eq 7.45 using K_v = b z^n with n = 1 - m (Ch.7 power-law setup).
    Inputs:
      - e1_mb, e2_mb: vapor pressure at heights z1_cm, z2_cm upstream (mb)
      - z1_cm, z2_cm: measurement heights (cm), z2_cm > z1_cm
      - u2_mps: wind speed at 2 m (m/s), used to infer u_* via power-law wind profile
    Returns:
      E_a in g cm^-2 s^-1  (multiply by 864000 for mm/day)
    """
    # specific humidity at the two heights (dimensionless)
    q1 = 0.622 * e1_mb / p_mb
    q2 = 0.622 * e2_mb / p_mb

    # infer u_* from U(z)=Cp*u_*(z/z0)^m at z=2 m = 200 cm
    z2ref_cm = 200.0
    u2_cms = 100.0 * u2_mps
    ustar_cms = u2_cms / (Cp * (z2ref_cm / z0_cm)**m)

    # Eq (7.55) coefficient b (so that K_v = b z^n with n=1-m)
    b = ustar_cms * (z0_cm**m) / (Cp * m)
    n = 1 - m

    Ea = rho * b * (1 - n) * (q1 - q2) / (z2_cm**(1 - n) - z1_cm**(1 - n))
    return Ea

# --- Plot for representative values ---
xf = np.logspace(0, 5, 100)  # fetch [m]: 1 m to 2 km

m = 1/7
z0_cm = 20     #  (representative "effective" roughness; adjust as needed)
de_mb = 15.0     # es - eas [mb] (e.g., wet surface near saturation at 
                 #    20°C vs drier upwind air)
e2_mb = 10
e1_mb = de_mb + e2_mb


for u2 in [2.0, 5.0, 8.0]:   # m/s
    enh = evap_enhancement_754(xf, u2, de_mb, m=m, z0_cm=z0_cm, Cp=5.5)
    enh_mmday = enh * 864000.0   # (g/cm^2/s) -> mm/day
    c =  Ea_from_745_powerlaw(u2, z0_cm, m,
                         e1_mb, e2_mb, z0_cm, 2000,
                         Cp=5.5, rho=1.2e-3, p_mb=1013.2) * 864000.0  
    plt.plot(xf, enh_mmday, label=f"$u2$ = {u2} m/s")
    
plt.xscale("log")
plt.xlabel("fetch $x_f$ (m)")
plt.ylabel(r"enhancement $(\bar E - E_a)$ (mm/day)")
plt.title(r"Eq. (7.54) leading-edge enhancement")
# (example: $m=1/7$, $\hat z_0=0.03$ cm, $\Delta e=15$ mb)
plt.grid(True, which="both", alpha=0.3)
plt.legend()
plt.show()

# adapt to temperature.
```
