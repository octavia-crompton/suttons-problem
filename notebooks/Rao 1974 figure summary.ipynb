### Boundary & Initial Conditions from Rao, Wyngaard & Coté (1974)

*Local Advection of Momentum, Heat, and Moisture in Micrometeorology*  

**Citation:** Rao, K. S.; Wyngaard, J. C.; Coté, O. R. (1974). *Boundary-Layer Meteorology* **7**, 331–348.

---

__Variable definitions__

* $E$ is absolute humidity (mass of water vapor per volume, kg m$^{-3}$); 
* $Q$ is specific humidity (mass of water vapor per mass of moist air, kg kg$^{-1}$).

    Direct relationship: $E = \rho Q$, where $\rho$ is the moist‑air density (kg m$^{-3}$).

    Via vapor pressure: $E = \dfrac{e}{R_v T}$ and $Q = \dfrac{0.622 e}{p - 0.378e}$.
    
    Combine to get $E = \dfrac{Q p}{(0.622 + 0.378Q) R_v T}$ 
    
    with $p$ total pressure, $T$ temperature, $R_v$ the vapor gas constant.

Surface values: 

$$E_{01}=6.34~\mathrm{g\ m^{-3}}$$
$$E_{02}=15.84~\mathrm{g\ m^{-3}}$$



__Surface energy balance__

   $$H_{02}(x) + L\,F_{02}(x) = R_N - G$$

The available energy at the surface ($R_N-G$) is partitioned into sensible and latent heat fluxes that warm the air and evaporate water. Over a wetter patch, more of $R_N-G$ typically goes into $LE$, leaving less for $H$ and cooling the surface layer.

__Surface humidity (RH) boundary condition__
  
  $$Q_0(x) = R_s Q_s \big(\Theta_0(x)\big)$$

The surface air right above the wet patch is assumed to be at a fixed relative humidity $R_s$.


Surface-layer (Monin–Obukhov) flux–profile relations:

$$
\frac{\partial \Theta}{\partial z}=-\frac{T_*}{k z}\phi_h,\qquad
\frac{\partial Q}{\partial z}=-\frac{q_*}{k z}\phi_q,
$$

$$
T_* = -\frac{\overline{w\theta}}{u_*},\qquad
q_* = -\frac{\overline{wq}}{u_*}
$$

$$
H = -\rho c_p\,\overline{w\theta},\quad LE = -\rho L\,\overline{wq},
$$

$$
\{\,\Theta_0(x),\,Q_0(x),\,H_{02}(x),\,F_{02}(x)\,\}
\quad\text{and hence}\quad
\{\,u_*(x),\,T_*(x),\,q_*(x)\,\}.
$$





**1) Lower Boundary (surface at $z=z_0$)**

**Kinematic & thermodynamic (Dirichlet)**
- $U=W=0$ at $z=z_0$ 
- Scalars at the wall: $\Theta=\Theta_0$, $Q=Q_0$ at $z=z_0$.

**Flux–profile relations**
- $\displaystyle \frac{\partial U}{\partial z}=\frac{u_*}{k z_0}$,  
- $\displaystyle \frac{\partial \Theta}{\partial z}=-\frac{T_*}{k z_0}$,  
- $\displaystyle \frac{\partial Q}{\partial z}=-\frac{q_*}{k z_0}$ at $z=z_0$.

**Turbulent fluxes**

- $\overline{uw}\big|_{z_0}=-u_*^2$.
- $\overline{w\theta}\big|_{z_0}=-T_*\,u_*$ so $H_0=-\rho c_p\,\overline{w\theta}$.
- $\overline{w q}\big|_{z_0}=-q_*\,u_*$ so $F_0=-\rho\,\overline{w q}$.

**Constants**
- von Kármán $k=0.35$; $\rho=1075\ \mathrm{g\,m^{-3}}$
- $c_p=0.24\ \mathrm{cal\,g^{-1}\,K^{-1}}$


__Surface scales & conversions__

- Shear stress: $\tau_0=\rho u_*^2$, $\overline{u'w'}|_{z_0}=-u_*^2$.
- Sensible heat flux: $H_0=-\rho c_p\,\overline{w\theta}$, with $T_*=\dfrac{H_0}{\rho c_p u_*}$.
- Moisture flux: $F_0=-\rho\,\overline{w q}$ and $q_*=\dfrac{F_0}{\rho u_*}$.
- von Kármán $k=0.35$; $\rho$, $c_p$ as above.




__Upstream Boundary (dry, smooth terrain; $x\le 0$)__

- Prescribed surface parameters: $z_0=z_{01}$, $u_*=u_{*1}$, $\Theta_0=\Theta_{01}$, $Q_0=Q_{01}$.
- Upstream flux scales: $T_*=\dfrac{H_{01}}{\rho c_p u_{*1}}$, $q_*=\dfrac{F_{01}}{\rho u_{*1}}$.
- Upstream mean humidity is height-independent $\Rightarrow F_{01}=0$ and $q_*=0$.

__Downstream Boundary (wet, rougher terrain; $x>0$)__

At each fetch $x>0$, solve for the surface triplet $\{u_*(x), \Theta_0(x), Q_0(x)\}$; wall conditions then follow.

**(a) Set roughness and starred scales**
- $z_0=z_{02}$; 
- Downstream flux scales: $T_*=\dfrac{H_{02}}{\rho c_p u_{*2}}$, $q_*=\dfrac{F_{01}}{\rho u_{*2}}$.


**(b) Surface energy balance**
- $\displaystyle H_{02}(x) + L\,F_{02}(x) = R_N - G$.
  - Base case: $R_N-G$ taken equal to the **upstream** sensible flux $H_{01}$.
  - Variants: consider larger $R_N-G$ over the wet patch.

**(c) Surface moisture (RH) condition**
- $\displaystyle Q_0(x) = R_s\,Q_s\!\big(\Theta_0(x)\big)$,  
  surface specific humidity equals a fixed relative humidity $R_s$ times saturation at the surface temperature.

**(d) Mean-form closure**
- Use flux–profile mean relations (their Eqs. 18–19) with (c) to solve for $u_*, \Theta_0, Q_0$ at each $x$.

**(e) Matching at the discontinuity $x=0$ (useful estimate)**
- $\displaystyle \frac{\Delta \Theta_0}{\Delta E_0} \approx \frac{L}{\rho c_p}$,  
  used with (c) to estimate $\Theta_0$, $Q_0$ immediately on the wet side.


#### 4) Initial Conditions for the march in $x$

- Build an equilibrium upstream column by applying upstream wall conditions until a steady, horizontally homogeneous state is reached.
- Use that upstream equilibrium column as the **inflow initial state** at the terrain discontinuity.

__Surface-layer “star” scales__


- **$u_*$ — friction velocity (m s$^{-1}$)**  
  Velocity scale set by surface stress:
  $
  u_* \equiv \sqrt{\frac{\tau_0}{\rho}} \;=\; \sqrt{-\,\overline{u'w'}}
  $
   - axes aligned with mean wind so $\overline{v'w'}\!\approx\!0$

- **$T_*$ — temperature (potential-temperature) scale (K)**  
  From the kinematic sensible-heat flux:
  $
  T_* \equiv -\,\frac{\overline{w\theta}}{u_*}
  $
  - Using the heat flux $H$ (W m$^{-2}$), with $H=-\rho c_p\,\overline{w\theta}$:
  $ T_*=\frac{H}{\rho c_p\,u_*}.$
  - Sign: upward sensible heat ($\overline{w\theta}>0$) $\Rightarrow T_*<0$.

- **$q_*$ — specific-humidity scale (kg kg$^{-1}$)**  
  From the kinematic moisture flux:
  $ q_* \equiv -\,\frac{\overline{wq}}{u_*}$
  With evaporation mass flux $E$ (kg m$^{-2}$ s$^{-1}$) or latent-heat flux $LE$ (W m$^{-2}$), $LE=L_v E$:
 $$q_*=\frac{E}{\rho\,u_*}$$ 
 $$q_*= \frac{LE}{\rho\,L_v\,u_*}.$$
  - Sign: upward moisture flux ($\overline{wq}>0$) $\Rightarrow q_*<0$.

\frac{\partial \Theta}{\partial z}=-\frac{T_*}{kz}\,\phi_h, \qquad
\frac{\partial Q}{\partial z}=-\frac{q_*}{kz}\,\phi_q,
$  

with von Kármán $k$, stability functions $\phi_\bullet$, and height $z$.

#### Why does $\dfrac{\Delta \Theta_0}{\Delta E_0} \approx \dfrac{L}{\rho\,c_p}$ “match” the discontinuity at $x=0$?

**Setting:** At $x=0$ the surface switches abruptly from dry (upwind) to wet (downwind). Just *above* the surface, the air has not yet advected/adjusted; what changes instantly are the **wall values** $(\Theta_0,Q_0)$ on the wet side.

 1) Start from the surface energy balance on both sides
Upwind and (immediately) downwind use
$
H + L\,F = R_N - G.
$
In the **base case** for Figs. 1–2, $R_N-G$ is taken the *same* across the line at $x=0$. Subtracting the two sides gives
$
\Delta(H + L\,F)\;=\;0 \quad\Longrightarrow\quad \Delta H + L\,\Delta F \;=\;0.
$

 2) Relate flux changes to *wall* scalar changes at $x=0$
Near the wall,
$
H \;=\; -\,\rho c_p\,\overline{w\theta} \;=\; \rho c_p\,u_*\,T_*,
\qquad
F \;=\; -\,\rho\,\overline{wq} \;=\; \rho\,u_*\,q_*,
$
and the (neutral/Kansas) flux–profile relations give, at a small reference height $z_1$,
$
T_* \approx -\,\frac{k\,$Theta(z_1)-\Theta_0]}{\ln(z_1/z_0)},\qquad
q_* \approx -\,\frac{k\,[Q(z_1)-Q_0]}{\ln(z_1/z_0)}.
$
Right at the discontinuity the air **just above** the surface is (to leading order) the same on both sides, so $\Theta(z_1)$ and $Q(z_1)$ are (nearly) unchanged; the *jumps* are in $\Theta_0$ and $Q_0$.

Hence the **changes** in the turbulent fluxes are proportional to the **changes** in the wall scalars:
$
\Delta H \;\propto\; -\,\rho c_p\,\Delta\Theta_0,\qquad
\Delta F \;\propto\; -\,\rho\,\Delta Q_0,
$
with the same positive proportionality factor (it cancels out).

Plugging into $\Delta H + L\,\Delta F=0$ gives
$
\rho c_p\,\Delta\Theta_0 + \rho L\,\Delta Q_0 \;\approx\; 0
\quad\Longrightarrow\quad
\frac{\Delta \Theta_0}{\Delta Q_0} \;\approx\; -\,\frac{L}{c_p}.
$

 3) Convert specific to absolute humidity
The paper often quotes **absolute humidity** $E$ (kg m$^{-3}$ or g m$^{-3}$). For small changes near the surface,
$
E \;\approx\; \rho\,Q \quad\Longrightarrow\quad \Delta E_0 \;\approx\; \rho\,\Delta Q_0.
$
Therefore
$
\boxed{\;\frac{\Delta \Theta_0}{\Delta E_0} \;\approx\; \frac{L}{\rho\,c_p}\;}
$
where the minus sign dropped because they use **magnitudes**: the wet side has $\Delta \Theta_0<0$ (cooler) and $\Delta E_0>0$ (moister).

 4) Physical meaning
- At $x=0$ the **available energy** $R_N-G$ is (initially) unchanged, so any **increase in latent heat** ($L\,\Delta F>0$) must be offset by an **equal decrease in sensible heat** ($\Delta H<0$).
- In the wall layer, $H$ is controlled by the **temperature drop** to the surface, and $F$ by the **humidity drop** to the surface. If the eddy transfer for heat and vapor is comparable, the required **ratio** of temperature drop to humidity drop is set by the **conversion factors** $c_p$ and $L$ (plus $\rho$ to switch $Q\leftrightarrow E$).
- That’s why this relation “**matches at the discontinuity**”: it provides the **consistent pair** $(\Theta_0,Q_0)$ on the wet side that keeps $H+LE$ continuous across $x=0$ when the surface type changes abruptly.

 5) How it’s used with the RH condition (their Eq. (17))
You combine the **matching slope** with the **wet‑patch surface RH**:
$
Q_0(0^+) = R_s\,Q_s\!\big(\Theta_0(0^+)\big),
\qquad
\Theta_0(0^+) \approx \Theta_{01} - \frac{L}{\rho c_p}\,\big(E_0(0^+) - E_{01}\big).
$
These two equations determine the **immediate wet‑side** wall state $\{\Theta_0(0^+),Q_0(0^+)\}$, which is then used as the starting point for integrating downstream.

 6) Quick numerical sanity check (using the paper’s constants)
With $L\approx 585~\mathrm{cal\,g^{-1}}$, $\rho\approx 1075~\mathrm{g\,m^{-3}}$, $c_p\approx 0.24~\mathrm{cal\,g^{-1}\,K^{-1}}$:
$
\frac{L}{\rho c_p}\;\approx\;\frac{585}{1075\times 0.24}\;\approx\;2.27~\mathrm{K}\,\Big/\,\frac{\mathrm{g}}{\mathrm{m^3}}.
$
If $\Delta E_0 \approx 9.5~\mathrm{g\,m^{-3}}$ (their figure/text),
$
|\Delta \Theta_0| \;\approx\; 2.27 \times 9.5 \;\approx\; 21.6~\mathrm{K},
$
which matches the reported $\Delta \Theta_0 \approx 21^\circ\mathrm{C}$ at $x=0$.

---

Step 2 — Relate $H$ and $F$ to wall scalar differences

From the wall relations (Eq. 13) and the mean‑form expressions (Eqs. 18–19), the turbulent surface fluxes can be written in terms of near‑surface mean–wall differences at some small reference height $z_1$:

* $ H = -  \rho c_p,\overline{w\theta} = \rho c_p \ u_* \ T_* \propto \rho c_p  \ \big[\Theta(z_1)-\Theta_0\big] $
* $ F = -  \rho \overline{wq} = \rho u_* q_* \propto \rho \ \big[Q(z_1)-Q_0\big]. $

Right at $x=0$ the air just above the surface is to leading order the same on both sides, so $\Theta(z_1)$ and $Q(z_1)$ are nearly unchanged; the jumps come from the wall values $\Theta_0$ and $Q_0$. 

Thus, to first order, the flux changes are proportional to the jumps in the wall scalars with the *same* aerodynamic factor (from $u_*$, $k$, $\ln(z_1/z_0)$, etc.), so it cancels in the ratio. 

So, write:
* $ \Delta H \approx -\rho c_p C \Delta\Theta_0, \qquad \Delta F \approx -\rho C \Delta Q_0 $

with the same positive transfer factor $C$ for heat and vapor at $x=0$ (hence their “approximately”). Insert these in $\Delta H + L \Delta F=0$ to get:

$ \rho c_p \Delta\Theta_0 + \rho L\Delta Q_0 \approx 0 \Rightarrow \displaystyle \frac{\Delta\Theta_0}{\Delta Q_0} \approx - \frac{L}{c_p}. $

---

Step 3 — Switch from specific to absolute humidity

Near the surface, $E \approx \rho Q$, so $\Delta E_0 \approx \rho \Delta Q_0$. 

Therefore
$ \displaystyle \frac{\Delta\Theta_0}{\Delta E_0} \approx \frac{L}{\rho c_p}. $

Eq. 20 followsfrom the mean field equations and the surface energy balance. 

---


Eq. 20 together with the wet‑surface RH condition $Q_0 = R_s \ Q_s(\Theta_0)$ (Eq. 17) is used to *estimate the immediate wet‑side* wall state $\Theta_{02}$ and $Q_{02}$ at the discontinuity, given the known upwind wall state. 
Solve Eqs. (17) and (20) for the wet‑side surface values at $x=0$. 



Assumptions in the “approximately”

* $R_N-G$ is the same on the two sides right at $x=0$ (their base case). 
* The same near‑wall transfer factor multiplies heat and vapor at $x=0$ (so it cancels in the ratio); any roughness/stability differences at the step are second‑order for this matching estimate. 
* Use of $E\approx \rho Q$ with $\rho$ treated constant near the surface.
* All wall values are defined at $z=z_0$ (their lower‑boundary location). 

In the paper’s own words, Eq. 20 is an **approximate** relation derived from the mean‑field and energy‑balance equations and is used specifically to set the *wet‑side* surface values at the discontinuity before integrating downstream. 



**Interpretation (Rao et al. Eq. 20)**

It refers to the immediate step at the discontinuity (($x=0^-$) vs. ($x=0^+$)), not far-downwind equilibria.

$$
\frac{\Delta \Theta_0}{\Delta E_0} \approx \frac{L}{\rho,c_p},.
$$


1. **Surface energy balance at ($x=0^+$):**
   $$
   H_{02} + L\,F_{02} = R_N - G \equiv H_{01},
   $$
   so
   $$
   L\,F_{02} = H_{01} - H_{02},.
   $$

2. **Near-surface flux–profile scalings:**
   $$
   H_0 = \rho,c_p,u_*,T_*, \qquad
   F_0 = \rho,u_*,q_*,
   $$
   with
   $$
   T_* \propto \Delta \Theta_0, \qquad q_* \propto \Delta E_0,.
   $$

3. **Eliminate the common transfer coefficient (same (u_*) at the wall):**
   $$
   \frac{H_0}{L,F_0}
   \approx
   \frac{\rho,c_p,\Delta \Theta_0}{L,\Delta E_0},.
   $$

4. **Evaluate at the discontinuity and combine with the balance above** to obtain the proportionality
   $$
   \frac{\Delta \Theta_0}{\Delta E_0}
   \approx
   \frac{L}{\rho,c_p},.
   $$

*Symbols:* (\Delta \Theta_0) = surface potential-temperature jump at (x=0) (\Delta E_0) = surface water-vapor (moisture) jump (L) = latent heat of vaporization (\rho) = air density (c_p) = specific heat of air at constant pressure (H) = sensible heat flux (F) = moisture flux.


```python
# Plot neutral log fits up to z = 10 m (linear z-axis) and include data markers
import numpy as np
import matplotlib.pyplot as plt

# ----- Anchors from the user's read-off of Fig. 1 -----
z1 = 0.01       # m
z2 = 10.0       # m
T1_x0  = 39.0   # °C at x = 0 m (dry)
T1_x20 = 28.5   # °C at x = 20 m (wet)
T2     = 30.0   # °C at z = 10 m (both)

# ----- Fit Theta(z) = a + b ln(z) from two points -----
def fit_log_two_points(z_a, T_a, z_b, T_b):
    ln_a, ln_b = np.log(z_a), np.log(z_b)
    b = (T_b - T_a) / (ln_b - ln_a)
    a = T_a - b * ln_a
    return a, b

a0,  b0  = fit_log_two_points(z1, T1_x0,  z2, T2)   # x=0 m (dry)
a20, b20 = fit_log_two_points(z1, T1_x20, z2, T2)   # x=20 m (wet)

# ----- Surface-layer equivalents: Theta = Theta0 - (T*/kappa) ln(z/z0) -----
kappa  = 0.35
z0_dry = 2.0e-5   # m  (Rao dry roughness)
z0_wet = 1.4e-3   # m  (Rao wet roughness)

def surface_params(a, b, z0, kappa):
    T_star = -kappa * b
    Theta0 = a - (T_star / kappa) * np.log(z0)  # since a = Theta0 + (T*/kappa) ln z0
    return Theta0, T_star

Theta0_dry, Tstar_dry = surface_params(a0,  b0,  z0_dry, kappa)
Theta0_wet, Tstar_wet = surface_params(a20, b20, z0_wet, kappa)

# ----- z-range for plotting (linear axis) -----
zmin = 0.005   # m, just above the larger roughness length
zmax = 10.0    # m
z = np.linspace(zmin, zmax, 600)

Theta_fit_x0  = a0  + b0  * np.log(z)
Theta_fit_x20 = a20 + b20 * np.log(z)

# ----- Plot with data markers included -----
plt.figure(figsize=(6.5, 5.5))
plt.plot(Theta_fit_x0,  z, label="Fit: x=0 m (dry)")
plt.plot(Theta_fit_x20, z, label="Fit: x=20 m (wet)")
# Data markers
plt.scatter([T1_x0, T2],  [z1, z2],  marker='o', label="Data: x=0 m")
plt.scatter([T1_x20, T2], [z1, z2], marker='s', label="Data: x=20 m")
plt.xlabel("Theta (°C)")
plt.ylabel("z (m)")
plt.title("Neutral log fits up to z = 10 m (with data markers)")
plt.ylim(0, zmax)
plt.legend(loc="lower right")
plt.tight_layout()
plt.show()

# ----- Print the equations (Theta in °C, z in m) -----
def eq_strings(a, b, Theta0, Tstar, kappa, z0):
    eq1 = f"Theta(z) = {a:.3f} + ({b:.6f}) * ln(z)"
    eq2 = f"Theta(z) = {Theta0:.2f} - ({Tstar:.4f}/{kappa:.2f}) * ln(z/{z0:.1e})"
    return eq1, eq2

eq1_dry,  eq2_dry  = eq_strings(a0,  b0,  Theta0_dry, Tstar_dry, kappa, z0_dry)
eq1_wet,  eq2_wet  = eq_strings(a20, b20, Theta0_wet, Tstar_wet, kappa, z0_wet)

print("Fitted equations (Theta in °C, z in m):")
print("  x=0 m  (dry):")
print("    ", eq1_dry)
print("    ", eq2_dry)
print("  x=20 m (wet):")
print("    ", eq1_wet)
print("    ", eq2_wet)
```
